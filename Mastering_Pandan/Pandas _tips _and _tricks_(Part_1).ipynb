# 03:How to  Rename the columns 

```python
df=pd.DataFrame({'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})
df
```

```python
# Renmaing the columns
df.rename(columns={'A':'a','B':'b','C':'c'},inplace=True)
```

```python
# Another way to rename the columns
df.columns=['a_col','b_col','c_col']
df
```

```python
# To replace the any character or string  in the column name
df.columns=df.columns.str.replace('_','')
```

# 05: Add Prefix in column

```python
# How to add prefix or suffix to the column name
df=df.add_prefix('X_')

df
```

```python
# we should use the add_suffix to add the suffix to the column name
df=df.add_suffix('_Y')
df
```

```python
# To Retun the colunms its orginal name
df.columns=['Col_1','Col_2','Col_3']
df
```
