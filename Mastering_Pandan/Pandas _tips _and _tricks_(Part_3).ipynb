# 15: Filtering  and Sorting 

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
# How to Check the unique  value
df.sex.unique()
df.embark_town.unique()
```

```python
# if we want to see only female data
df[(df.sex == 'female')]
```

```python
df.who.unique()
```

```python
df.age.unique()
```

```python
df.embark_town.unique()
```

```python
# if we can see only the data of the people of Female  who are from Southampton 
df[(df.embark_town=='Southampton') & (df.sex=='female' )]
```

```python
# if we want to see the data of the people of female  who are from Southampton and as well as from Queenstown
df[((df.embark_town=='Southampton') | (df.embark_town=='Queenstown')) &(df.sex=='female' )]
```

```python
# if we want see those people who are from southamptan as well as from Queenstown
df[df.embark_town.isin(['Southampton','Queenstown'])]
```

```python
# i we want to see those peaople who are not greater than 30
df[df.age >30]
```
