# **Earth Engine data types**

In Google Earth Engine, there are several data types that you can work with. Here are some of the main ones:

**1 Image:** Represents raster data, such as ***satellite imagery***, where each pixel has a value.

**2. Image Collection:** A collection of images, often from the same source or sensor, allowing you to work with ***multiple images*** simultaneously.

**3.Features**: Represents ***vector data***, such as points, lines, or polygons. Features can have properties associated with them, like ***attributes*** in a GIS.

**4.Feature Collection**: A collection of features, allowing you to work with ***multiple features*** simultaneously.

**5.Geometry**: Represents ***geometric objects*** like points, lines, or polygons, which can be used for ***spatial operations***.

# **Filtering image collections**
- Filtering image collections involves the process of **refining** a set of images based on **specific criteria** to meet particular requirements or preferences.Some Basic Filtering Image function Below:<br>
**Basic Image Filtering Functions:**

1. **Filter by Date:**
   Narrow down images based on acquisition dates to focus on specific time periods.
   
2. **Filter by Cloud Cover:**
   Exclude images with high cloud cover percentages for clearer and more accurate results.

3. **Filter by Region of Interest (ROI):**
   Limit images to a defined geographical area of interest to tailor analysis to specific locations.

4. **Filter by Sensor Characteristics:**
   Select images captured by specific sensors or platforms to meet project requirements.

5. **Filter by Image Quality Metrics:**
   Refine image selection based on quality metrics such as resolution, signal-to-noise ratio, or geometric accuracy.

6. **Filter by Band Combination:**
   Choose images with desired spectral bands for customized analysis and visualization.

7. **Filter by Metadata Attributes:**
   Utilize metadata attributes like sun elevation, azimuth angles, or sensor viewing angles for precise image selection.

```python
# This imagery have cloud cover, so thatwhy is not clear visulization
Map = geemap.Map()
collection = (
    ee.ImageCollection('COPERNICUS/S2_SR') # Load Sentinel images collection
    .filterDate('2021-01-01', '2022-01-01') # Filter images by date

)
image = collection.median() # Take median

vis = {
    'min': 0.0,
    'max': 3000,
    'bands': ['B4', 'B3', 'B2'], # True color visulization
}

Map.setCenter(83.277, 17.7009, 12)
Map.addLayer(image, vis, 'Sentinel-2')
Map
```

#### **Note:**Cloud cover in the imagery may affect the ***clarity*** of visualization, especially in areas with ***significant cloud presence***. Adjustments to visualization parameters or additional preprocessing steps may be necessary to ***improve visualization*** clarity in such cases.

```python
# Let load cloud free imagey
Map = geemap.Map()
collection = (
    ee.ImageCollection('COPERNICUS/S2_SR') # Load Sentinel images collection
    .filterDate('2021-01-01', '2022-01-01') # Filter images by date
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 5))
)
image = collection.median() # Take median

vis = {
    'min': 0.0,
    'max': 3000,
    'bands': ['B8', 'B4', 'B3'], # Fasle color visulization
}

Map.setCenter(83.277, 17.7009, 12)
Map.addLayer(image, vis, 'Sentinel-2')
Map
```

#### **Note:** The median method for imagery visualization helps to reduce the impact of outliers, such as clouds, atmospheric effects, and sensor noise, resulting in a clearer representation of the underlying landscape's typical spectral characteristics

# **Geometry & Its Types**
In geemap, **geometry** refers to spatial objects such as points, lines, and polygons used to represent geographic features.

The main types of geometries in geemap include:

1. **Point**: A single location defined by its coordinates (latitude and longitude).
2. **LineString**: A sequence of connected line segments, representing linear features like roads or rivers.
3. **Polygon**: A closed shape with three or more vertices, representing areas like countries, lakes, or buildings.

```python
# Let create polygone and rectangle
Map = geemap.Map()

# Define a rectangle geometry with bounding box coordinates
rectangle = ee.Geometry.Rectangle([-40, -20, 40, 20], None, False)

# Define a polygon geometry with vertices at specified coordinates
polygon = ee.Geometry.Polygon(
    [[[-5, 40], [65, 40], [65, 60], [-5, 60], [-5, 60]]]
)

# Add the rectangle and polygon layers to the map
Map.addLayer(rectangle, {}, 'Rectangle')
Map.addLayer(polygon, {}, 'Polygon')

# Display the map
Map
```

```python
Map = geemap.Map()

# Define a point geometry at coordinates (1.5, 1.5)
point = ee.Geometry.Point([1.5, 1.5])

# Define a line string geometry with vertices at specified coordinates
lineString = ee.Geometry.LineString(
    [[-35, -10], [35, -10], [35, 10], [-35, 10]], None, False
)

# Define a linear ring geometry with vertices at specified coordinates
linearRing = ee.Geometry.LinearRing(
    [[-35, -10], [35, -10], [35, 10], [-35, 10], [-35, -10]], None, False
)

# Define a rectangle geometry with bounding box coordinates
rectangle = ee.Geometry.Rectangle([-40, -20, 40, 20], None, False)

# Define a polygon geometry with vertices at specified coordinates
polygon = ee.Geometry.Polygon(
    [[[-5, 40], [65, 40], [65, 60], [-5, 60], [-5, 60]]], None, False
)

# Add each geometry layer to the map
Map.addLayer(point, {}, 'Point')
Map.addLayer(lineString, {}, 'LineString')
Map.addLayer(linearRing, {}, 'LinearRing')
Map.addLayer(rectangle, {}, 'Rectangle')
Map.addLayer(polygon, {}, 'Polygon')

# Display the map
Map
```
