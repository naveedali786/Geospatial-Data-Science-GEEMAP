# **Earth Engine data types**

In Google Earth Engine, there are several data types that you can work with. Here are some of the main ones:

**1 Image:** Represents raster data, such as ***satellite imagery***, where each pixel has a value.

**2. Image Collection:** A collection of images, often from the same source or sensor, allowing you to work with ***multiple images*** simultaneously.

**3.Features**: Represents ***vector data***, such as points, lines, or polygons. Features can have properties associated with them, like ***attributes*** in a GIS.

**4.Feature Collection**: A collection of features, allowing you to work with ***multiple features*** simultaneously.

**5.Geometry**: Represents ***geometric objects*** like points, lines, or polygons, which can be used for ***spatial operations***.

# **Visualizing Earth Engine Single Band images**

```python
# Create a map
map = geemap.Map()

# Load the AHN Netherlands 0.5m DEM, Interpolated data
image = ee.Image('AHN/AHN2_05M_INT')

# Define visualization parameters
vis_params = {
    'min': -0.5,
    'max': 30.0,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5']
}

# Add the image to the map
map.addLayer(image, vis_params, 'DEM_0.5')

# Display the map
map
```

# **Loading Cloud GeoTIFFs**

```python
Map = geemap.Map()
URL = 'https://bit.ly/3aSZ0fH'
image = geemap.load_GeoTIFF(URL)
vis = {
    "min": 3000,
    "max": 13500,
    "bands": ["B3", "B2", "B1"],
}
Map.addLayer(image, vis, 'Cloud GeoTIFF')
Map.centerObject(image)
Map
```
